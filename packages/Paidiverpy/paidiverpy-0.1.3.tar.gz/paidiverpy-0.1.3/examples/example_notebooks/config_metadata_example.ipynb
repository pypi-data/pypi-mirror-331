## Open config file and metadata

This example shows how to handle and parse metadata and configuration files using paidiverpy package

## Import dependencies

```python
%load_ext autoreload
%autoreload 2
from paidiverpy import Paidiverpy
from paidiverpy.config.config import Configuration
from paidiverpy.metadata_parser import MetadataParser
from paidiverpy.open_layer import OpenLayer
```

## Loading metadata and configuration files

There are several ways to load and parse the metadata and the configuration file:

- Option 1: using the Paidiverpy class
- Option 2: using other processing classes, like OpenLayer#
- Option 3: using Configuration class and MetadataParser class

### Option 1: Using Paidiverpy class

```python
# Instantiate the class
# You may need to change the parameters on the config_simple.yml file for the correct paths on your computer
paidiverpy = Paidiverpy(config_file_path="../config_files/config_simple.yml")
```

```python
# See the type of config
type(paidiverpy.config)
```

```python
# Print config as json
paidiverpy.config
```

```python
# See metadata information
paidiverpy.get_metadata()
```

### Option 2: Using other processing classes, like OpenLayer

```python
open_layer = OpenLayer(config_file_path="../config_files/config_simple.yml")
```

```python
# Print config as json
open_layer.config
```

```python
# See metadata information
# The metadata in this case may have more information because the
# openlayer class extracts the exif information from the files
open_layer.get_metadata()
```

### Option 3: using Configuration class and MetadataParser class

#### Using configuration parsing

```python
config = Configuration(config_file_path="../config_files/config_simple.yml")
config
```


In this case, yu will not have access to the metadata, as it was not parsed yet

#### Parse only the metadata

```python
# You need to pass the configuration file
metadata = MetadataParser(config=config)
metadata
```
