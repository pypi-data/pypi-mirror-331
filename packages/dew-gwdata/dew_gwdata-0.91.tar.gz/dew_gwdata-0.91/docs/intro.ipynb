```python
import dew_gwdata as gd
db = gd.sageodata()
```

In the background this is using a different Python package called ``sageodata_db``, but you don't have to worry about that, you can access everything you need through here.

The object ``db`` is an SAGeodataConnection object. It will silently time out after about 6 hours so you will need to re-create it using the line above.

First some basics. You can, if you want, run SQL and get a result in the form of a pandas DataFrame

```python
db.query("""
select drillhole_no, dh_name
from dhdb.dd_drillhole_vw
where map_100000_no = 6641
and water_well_class = 'Y'
""")
```

```python
wells = db.find_wells("SLE69, YAT115 PTA41 6627-6983")
```

```python
wells
```

This returns just a list of wells. We can use that to fetch summary info from SA Geodata:

```python
df = db.wells_summary(wells)
df
```

``wells_summary()`` is what I can a "predefined query". It is basically one of 30-40 methods on ``SAGeodataConnection`` which return data from SA Geodata. Most of them take one query parameter which essentially means "a list of drillholes". Technically this means either a list of drillhole numbers, or a pandas DataFrame with a column called "dh_no". If you have a list of unit numbers or obs numbers instead, you have to use a different method first to get the drillhole numbers for those unit numbers etc. We'll cover that shortly.

The following pages of the documentation will cover all the different predefined queries that exist, which cover almost all the relevant data in the database.
